## **1. Carga y Exploración Inicial de Datos**

```python
import pandas as pd

# Cargar el conjunto de datos desde el archivo CSV
df = pd.read_csv('/content/Electric_Vehicle_Population_Data - Electric_Vehicle_Population_Data.csv.csv')

# Mostrar las primeras 5 filas del DataFrame
print("Primeras 5 filas del DataFrame:")
display(df.head())
```

```python
# Mostrar información general del DataFrame
print("\nInformación general del DataFrame:")
df.info()
```

```python
# Mostrar estadísticas descriptivas de las columnas numéricas
print("\nEstadísticas descriptivas del DataFrame:")
display(df.describe())
```

## **2. Limpieza y Preprocesamiento de Datos**

```python
# Contar el número de valores nulos por columna
print("\nNúmero de valores nulos por columna:")
print(df.isnull().sum())
```

```python
# Identificar y manejar valores '0' en 'Electric Range' y 'Base MSRP'
# Por ahora, simplemente contamos cuántos hay. Más adelante se decidirá la estrategia.
print("\nNúmero de valores '0' en 'Electric Range':", (df['Electric Range'] == 0).sum())
print("Número de valores '0' en 'Base MSRP':", (df['Base MSRP'] == 0).sum())

# Para las columnas con valores nulos, una estrategia inicial es eliminarlos o imputarlos.
# Dada la cantidad, optaremos por eliminar las filas con nulos en columnas clave como 'City' y 'County'
# para simplificar el análisis inicial, o imputarlos si tienen un impacto menor.
# Vamos a imputar los valores nulos en 'Legislative District' con la moda (valor más frecuente).
# Y para 'Vehicle Location', 'Electric Utility', '2020 Census Tract' con un valor 'Desconocido' o similar.

# Imputar 'Legislative District' con la moda
df['Legislative District'].fillna(df['Legislative District'].mode()[0], inplace=True)

# Imputar 'County', 'City', 'Postal Code', 'Vehicle Location', 'Electric Utility', '2020 Census Tract' con 'Desconocido' o la moda
# Para 'Postal Code' y '2020 Census Tract' que son numéricos y tienen pocos nulos, se podría usar la moda o la media/mediana.
# Optaremos por la moda por ahora para mantener la consistencia con distritos legislativos.
df['County'].fillna(df['County'].mode()[0], inplace=True)
df['City'].fillna(df['City'].mode()[0], inplace=True)
df['Postal Code'].fillna(df['Postal Code'].mode()[0], inplace=True)
df['Vehicle Location'].fillna('Desconocido', inplace=True)
df['Electric Utility'].fillna('Desconocido', inplace=True)
df['2020 Census Tract'].fillna(df['2020 Census Tract'].mode()[0], inplace=True)


print("\nValores nulos después de la imputación:")
print(df.isnull().sum())
```

## **3. Ingeniería de Características**

```python
# Para el propósito de 'Vehicles Per Capita', necesitaremos datos de población, que no están directamente en el dataset.
# Sin embargo, podemos aproximarlo usando '2020 Census Tract' y agrupando por ubicaciones más grandes como 'City' o 'County'.
# Por ahora, nos centraremos en características directamente extraíbles o calculables.

# Crear una nueva característica 'Vehicles Per County' o 'Vehicles Per City' para ver la concentración.
# Agruparemos por 'County' y 'City' y contaremos el número de vehículos.
vehiculos_por_condado = df.groupby('County')['VIN (1-10)'].count().reset_index()
vehiculos_por_condado.rename(columns={'VIN (1-10)': 'Total Vehiculos'}, inplace=True)
print("\nTop 10 Condados con más vehículos eléctricos:")
display(vehiculos_por_condado.sort_values(by='Total Vehiculos', ascending=False).head(10))

# Extraer la marca (Make) si ya no es una columna independiente, aunque en este dataset ya lo es.
# Asumiremos que 'Make' ya es la marca.
# Si hubiera necesidad de limpiar o estandarizar, se haría aquí.
print("\nTop 10 Marcas de vehículos eléctricos:")
print(df['Make'].value_counts().head(10))

# Crear una característica para la antigüedad del vehículo (si el año actual es relevante).
# Usaremos un año de referencia, por ejemplo, 2024 (asumiendo que es el año actual para el análisis).
año_actual = 2024
df['Antiguedad Vehiculo'] = año_actual - df['Model Year']
print("\nDistribución de la antigüedad de los vehículos:")
display(df['Antiguedad Vehiculo'].describe())
```

## **4. Análisis Exploratorio de Datos (EDA)**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Establecer un estilo para los gráficos
sns.set_style("whitegrid")

# 1. Distribución de vehículos eléctricos por tipo
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='Electric Vehicle Type', order=df['Electric Vehicle Type'].value_counts().index)
plt.title('Distribución de Vehículos Eléctricos por Tipo')
plt.xlabel('Número de Vehículos')
plt.ylabel('Tipo de Vehículo Eléctrico')
plt.show()

# 2. Top 10 de marcas de vehículos eléctricos
plt.figure(figsize=(12, 7))
sns.countplot(data=df, y='Make', order=df['Make'].value_counts().head(10).index, palette='viridis')
plt.title('Top 10 Marcas de Vehículos Eléctricos')
plt.xlabel('Número de Vehículos')
plt.ylabel('Marca')
plt.show()

# 3. Distribución de vehículos eléctricos por año de modelo
plt.figure(figsize=(14, 7))
sns.countplot(data=df, x='Model Year', palette='cubehelix')
plt.title('Distribución de Vehículos Eléctricos por Año de Modelo')
plt.xlabel('Año del Modelo')
plt.ylabel('Número de Vehículos')
plt.xticks(rotation=45)
plt.show()

# 4. Distribución de la autonomía eléctrica (Electric Range)
plt.figure(figsize=(12, 7))
sns.histplot(df['Electric Range'][df['Electric Range'] > 0], bins=50, kde=True, color='skyblue')
plt.title('Distribución de la Autonomía Eléctrica (excluyendo 0)')
plt.xlabel('Autonomía Eléctrica (millas)')
plt.ylabel('Frecuencia')
plt.show()

# 5. Distribución de Base MSRP (excluyendo 0)
plt.figure(figsize=(12, 7))
sns.histplot(df['Base MSRP'][df['Base MSRP'] > 0], bins=50, kde=True, color='lightcoral')
plt.title('Distribución de Base MSRP (excluyendo 0)')
plt.xlabel('Precio Base Sugerido por el Fabricante (MSRP)')
plt.ylabel('Frecuencia')
plt.xlim(0, 150000) # Limitar el eje X para una mejor visualización si hay valores atípicos muy grandes
plt.show()
```

## **5. Modelado Analítico**

```python
# Para el modelado analítico, podemos enfocarnos en predecir ciertos aspectos o entender relaciones clave.
# Una pregunta interesante podría ser: ¿Qué factores influyen en la autonomía eléctrica de un vehículo?
# Sin embargo, dado que el objetivo es el análisis de tamaño de mercado y predicciones de mercado,
# nos centraremos en tendencias a lo largo del tiempo o factores que influyen en la adopción.

# Analizaremos la tendencia de crecimiento del mercado por año de modelo.
# Esto nos dará una base para posibles proyecciones futuras.

vehiculos_por_año = df.groupby('Model Year').size().reset_index(name='Número de Vehículos')
print("\nNúmero de vehículos eléctricos por año de modelo:")
display(vehiculos_por_año.tail(10))

# Visualizar la tendencia de crecimiento
plt.figure(figsize=(12, 6))
sns.lineplot(data=vehiculos_por_año, x='Model Year', y='Número de Vehículos', marker='o')
plt.title('Crecimiento del Mercado de Vehículos Eléctricos por Año de Modelo')
plt.xlabel('Año del Modelo')
plt.ylabel('Número de Vehículos Registrados')
plt.grid(True)
plt.show()

# Podemos intentar una regresión lineal simple para proyectar el crecimiento si la tendencia es clara.
# Filtramos los años más recientes para una proyección más relevante.
recent_years = vehiculos_por_año[vehiculos_por_año['Model Year'] >= 2010]

from sklearn.linear_model import LinearRegression
import numpy as np

X = recent_years['Model Year'].values.reshape(-1, 1)
y = recent_years['Número de Vehículos'].values

model = LinearRegression()
model.fit(X, y)

# Predecir para los próximos 5 años (hasta 2029 si el último año es 2024)
future_years = np.array(range(recent_years['Model Year'].max() + 1, recent_years['Model Year'].max() + 6)).reshape(-1, 1)
future_predictions = model.predict(future_years)

# Crear un DataFrame para los resultados de la predicción
predictions_df = pd.DataFrame({
    'Model Year': future_years.flatten(),
    'Número de Vehículos Predichos': future_predictions.round().astype(int)
})

print("\nPredicciones de crecimiento del mercado para los próximos 5 años:")
display(predictions_df)

# Visualizar la tendencia y las predicciones
plt.figure(figsize=(14, 7))
sns.lineplot(data=vehiculos_por_año, x='Model Year', y='Número de Vehículos', marker='o', label='Datos Históricos')
sns.lineplot(x=future_years.flatten(), y=future_predictions, marker='x', color='red', linestyle='--', label='Predicciones')
plt.title('Crecimiento Histórico y Predicción del Mercado de Vehículos Eléctricos')
plt.xlabel('Año del Modelo')
plt.ylabel('Número de Vehículos Registrados')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Combinar datos históricos y predichos para la visualización
# Asegurémonos de que 'vehiculos_por_año' tiene los últimos datos históricos

# Crear una copia de vehiculos_por_año para no modificar el original
combined_data = vehiculos_por_año.copy()

# Convertir 'Número de Vehículos' en predictions_df a int para que coincida con combined_data
predictions_df_copy = predictions_df.copy()
predictions_df_copy['Número de Vehículos Predichos'] = predictions_df_copy['Número de Vehículos Predichos'].astype(int)
predictions_df_copy.rename(columns={'Número de Vehículos Predichos': 'Número de Vehículos'}, inplace=True)

# Añadir una columna para identificar si es dato histórico o predicho
combined_data['Tipo'] = 'Histórico'
predictions_df_copy['Tipo'] = 'Predicción'

# Unir los DataFrames
combined_for_bar_chart = pd.concat([combined_data, predictions_df_copy], ignore_index=True)

# Filtrar los años para que el gráfico sea más legible, centrándose en la tendencia reciente y las predicciones
# Empezaremos desde un año donde la tendencia de crecimiento sea más clara, por ejemplo, 2010.
combined_for_bar_chart = combined_for_bar_chart[combined_for_bar_chart['Model Year'] >= 2010]

plt.figure(figsize=(15, 8))
sns.barplot(x='Model Year', y='Número de Vehículos', hue='Tipo', data=combined_for_bar_chart, palette={'Histórico': 'skyblue', 'Predicción': 'salmon'})
plt.title('Comparación de Vehículos Eléctricos Históricos y Predichos por Año de Modelo')
plt.xlabel('Año del Modelo')
plt.ylabel('Número de Vehículos Registrados')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## **6. Conclusiones y Recomendaciones Estratégicas**

Basándonos en el análisis realizado, podemos extraer las siguientes conclusiones y ofrecer recomendaciones estratégicas para las empresas interesadas en el mercado de vehículos eléctricos:

### **Conclusiones Clave:**

1.  **Crecimiento Exponencial del Mercado:** Los datos históricos y las predicciones futuras muestran un crecimiento robusto y acelerado en la adopción de vehículos eléctricos. El 'Año del Modelo' es un fuerte indicador de esta tendencia, con un aumento significativo de registros en los últimos años.
2.  **Concentración Geográfica:** Condados como King, Snohomish y Pierce lideran en la concentración de vehículos eléctricos, lo que sugiere que las estrategias de mercado podrían beneficiarse de un enfoque regional.
3.  **Dominio de Marcas Establecidas:** Marcas como Tesla mantienen una cuota de mercado dominante, pero hay espacio para otros fabricantes con un crecimiento constante.
4.  **Autonomía y Precios:** Una gran proporción de vehículos tiene una autonomía eléctrica de 0, lo que podría indicar la presencia de vehículos híbridos enchufables (PHEV) con un rango eléctrico limitado o datos no disponibles. La distribución de los MSRP base (excluyendo ceros) muestra una variedad de puntos de precio, pero con una concentración en ciertos segmentos.
5.  **Antigüedad del Parque Vehicular:** La mayoría de los vehículos eléctricos en circulación son relativamente nuevos (promedio de 3.5 años), lo que indica un mercado en expansión y una rápida rotación de modelos.

### **Recomendaciones Estratégicas:**

1.  **Enfoque en Segmentos de Mercado en Crecimiento:** Identificar y dirigirse a los segmentos de consumidores que están impulsando el crecimiento más rápido (ej. compradores de modelos más recientes) y aquellos en áreas geográficas de alta concentración.
2.  **Desarrollo de Infraestructura de Carga:** El crecimiento sostenido del mercado de vehículos eléctricos dependerá en gran medida de la expansión de la infraestructura de carga. Las empresas que inviertan o se asocien en este ámbito podrían obtener una ventaja competitiva.
3.  **Diversificación de Modelos y Precios:** Aunque Tesla domina, existe una demanda para una variedad de modelos y rangos de precios. Ofrecer opciones competitivas en diferentes segmentos (desde vehículos económicos hasta de lujo) puede captar una porción más amplia del mercado.
4.  **Innovación en Autonomía y Baterías:** Para vehículos con 'Electric Range' en 0, es crucial destacar los beneficios de los PHEV o la necesidad de una mayor autonomía en los BEV. Continuar mejorando la autonomía y reduciendo los costos de las baterías será clave para la adopción masiva.
5.  **Marketing Dirigido y Educación al Consumidor:** Aclarar las ventajas de los vehículos eléctricos (beneficios ambientales, ahorros a largo plazo, incentivos gubernamentales) puede ayudar a superar las barreras de adopción, especialmente en regiones con menor penetración.
6.  **Monitoreo Continuo del Mercado:** Dada la naturaleza dinámica del mercado de vehículos eléctricos, es fundamental realizar un seguimiento continuo de las tendencias, los avances tecnológicos y las políticas gubernamentales para ajustar las estrategias de manera proactiva.

### **Consideraciones para Futuros Análisis:**

*   **Datos de Población:** Para un análisis más preciso de 'Vehículos por Cápita', sería beneficioso integrar datos de población por condado o por código postal.
*   **Análisis de Sentimiento:** Investigar las opiniones y percepciones de los consumidores sobre los vehículos eléctricos podría ofrecer información valiosa sobre las barreras y los motivadores de compra.
*   **Impacto de Políticas Gubernamentales:** Analizar cómo las regulaciones y los incentivos gubernamentales afectan la adopción de vehículos eléctricos en diferentes regiones.
