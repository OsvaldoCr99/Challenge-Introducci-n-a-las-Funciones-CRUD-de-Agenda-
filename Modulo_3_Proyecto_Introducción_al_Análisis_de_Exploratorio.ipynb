## RecopilaciÃ³n de datos

```python
import pandas as pd

# Cargar el archivo CSV en un DataFrame de pandas, especificando la codificaciÃ³n y el separador
df = pd.read_csv('/content/Book 4(Hoja1).csv', encoding='latin1', sep=';')

# Mostrar las primeras 5 filas del DataFrame
display(df.head())
```

## Limpieza y ExploraciÃ³n de datos

```python
# Mostrar informaciÃ³n general del DataFrame (tipos de datos, valores no nulos)
df.info()

# Mostrar estadÃ­sticas descriptivas de las columnas numÃ©ricas
display(df.describe())
```

## VisualizaciÃ³n de datos

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Calcular mÃ©tricas de ataque y eficacia
# Evitar divisiÃ³n por cero, rellenando con NaN y luego 0 si es necesario
df['GF_per_Match'] = df['GF'] / df['Mp']
df['Conversion_Rate_GF_ShT'] = np.where(df['Sh T'] > 0, df['GF'] / df['Sh T'], 0)

# Mostrar los 10 mejores equipos por Goles por Partido
print("\nTop 10 equipos por Goles por Partido:")
display(df.sort_values(by='GF_per_Match', ascending=False).head(10)[['Team', 'GF_per_Match', 'Sh T', 'GF']])

# Mostrar los 10 mejores equipos por Tasa de ConversiÃ³n de Goles (Goles por Tiro Total)
print("\nTop 10 equipos por Tasa de ConversiÃ³n de Goles (Goles por Tiro Total):")
display(df.sort_values(by='Conversion_Rate_GF_ShT', ascending=False).head(10)[['Team', 'Conversion_Rate_GF_ShT', 'Sh T', 'GF']])
```

```python
# VisualizaciÃ³n de los 10 mejores equipos por Goles por Partido
plt.figure(figsize=(12, 6))
sns.barplot(x='Team', y='GF_per_Match', data=df.nlargest(10, 'GF_per_Match'), palette='viridis', hue='Team', legend=False)
plt.title('Top 10 Equipos por Goles por Partido')
plt.xlabel('Equipo')
plt.ylabel('Goles por Partido')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# VisualizaciÃ³n de los 10 mejores equipos por Tasa de ConversiÃ³n de Goles
plt.figure(figsize=(12, 6))
sns.barplot(x='Team', y='Conversion_Rate_GF_ShT', data=df.nlargest(10, 'Conversion_Rate_GF_ShT'), palette='magma', hue='Team', legend=False)
plt.title('Top 10 Equipos por Tasa de ConversiÃ³n de Goles (GF / Sh T)')
plt.xlabel('Equipo')
plt.ylabel('Tasa de ConversiÃ³n de Goles')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
