### **Menú CRUD: Gestión de Equipos y Jugadores de Fútbol**

Vamos a construir un sistema simple para gestionar información de equipos y jugadores. Este menú interactivo permitirá realizar las operaciones básicas de Crear, Leer (Consultar y Ver todos), Actualizar y Eliminar registros.

```python
import uuid # Importar el módulo uuid para generar IDs únicos

# Diccionario para almacenar los equipos (simulando una base de datos)
equipos = {}

# --- Funciones CRUD (placeholders por ahora) ---

def ver_todos_los_registros():
    print("\n--- Ver Todos los Equipos ---")
    if not equipos:
        print("No hay equipos registrados aún.")
    else:
        print(f"Hay {len(equipos)} equipo(s) registrado(s).")
        for id_equipo, datos_equipo in equipos.items():
            print(f"ID: {id_equipo}, Nombre: {datos_equipo['nombre']}, País: {datos_equipo['pais']}")
    print("------------------------------")

def crear_nuevo_registro():
    print("\n--- Crear Nuevo Equipo ---")
    nombre = input("Ingrese el nombre del equipo: ")
    pais = input("Ingrese el país del equipo: ")

    # Generar un ID único para el equipo
    id_equipo = str(uuid.uuid4())

    equipos[id_equipo] = {
        'nombre': nombre,
        'pais': pais
    }
    print(f"Equipo '{nombre}' de '{pais}' con ID '{id_equipo}' creado exitosamente.")
    print("--------------------------")


def consultar_registro():
    print("\n--- Consultar Equipo ---")
    criterio = input("Ingrese el ID o el nombre del equipo a consultar: ")
    equipo_encontrado = False

    for id_equipo, datos_equipo in equipos.items():
        if id_equipo == criterio or datos_equipo['nombre'].lower() == criterio.lower():
            print(f"\nID: {id_equipo}")
            print(f"Nombre: {datos_equipo['nombre']}")
            print(f"País: {datos_equipo['pais']}")
            equipo_encontrado = True
            break # Asumimos que el ID es único, o que solo queremos el primero que coincida por nombre

    if not equipo_encontrado:
        print(f"No se encontró ningún equipo con el ID o nombre '{criterio}'.")
    print("------------------------")


def actualizar_registro():
    print("\n--- Actualizar Equipo ---")
    id_a_actualizar = input("Ingrese el ID del equipo a actualizar: ")

    if id_a_actualizar in equipos:
        print(f"Equipo actual: Nombre: {equipos[id_a_actualizar]['nombre']}, País: {equipos[id_a_actualizar]['pais']}")
        nuevo_nombre = input(f"Ingrese el nuevo nombre para el equipo (deje en blanco para mantener '{equipos[id_a_actualizar]['nombre']}'): ")
        nuevo_pais = input(f"Ingrese el nuevo país para el equipo (deje en blanco para mantener '{equipos[id_a_actualizar]['pais']}'): ")

        if nuevo_nombre:
            equipos[id_a_actualizar]['nombre'] = nuevo_nombre
        if nuevo_pais:
            equipos[id_a_actualizar]['pais'] = nuevo_pais
        print(f"Equipo con ID '{id_a_actualizar}' actualizado exitosamente.")
    else:
        print(f"No se encontró ningún equipo con el ID '{id_a_actualizar}'.")
    print("-------------------------")


def eliminar_registro():
    print("\n--- Eliminar Equipo ---")
    id_a_eliminar = input("Ingrese el ID del equipo a eliminar: ")

    if id_a_eliminar in equipos:
        nombre_equipo = equipos[id_a_eliminar]['nombre']
        del equipos[id_a_eliminar]
        print(f"Equipo '{nombre_equipo}' con ID '{id_a_eliminar}' eliminado exitosamente.")
    else:
        print(f"No se encontró ningún equipo con el ID '{id_a_eliminar}'.")
    print("-----------------------")


# --- Menú Principal ---
def mostrar_menu():
    print("\n===== MENÚ DE GESTIÓN DE EQUIPOS =====")
    print("1. Ver todos los equipos")
    print("2. Crear nuevo equipo")
    print("3. Consultar equipo")
    print("4. Actualizar equipo")
    print("5. Eliminar equipo")
    print("6. Salir del programa")
    print("====================================")

def ejecutar_menu():
    while True:
        mostrar_menu()
        opcion = input("Seleccione una opción: ")

        if opcion == '1':
            ver_todos_los_registros()
        elif opcion == '2':
            crear_nuevo_registro()
        elif opcion == '3':
            consultar_registro()
        elif opcion == '4':
            actualizar_registro()
        elif opcion == '5':
            eliminar_registro()
        elif opcion == '6':
            print("¡Hasta luego!")
            break
        else:
            print("Opción no válida. Intente de nuevo.")

# Iniciar el programa
ejecutar_menu()
```
