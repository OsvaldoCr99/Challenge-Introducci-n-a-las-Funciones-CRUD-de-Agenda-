## Objetivo:
Detectar los tumores malignos y beningnos mediante el dataset que recaba múltiples características

1. Aplicar el algoritmo de KNN a ([data.csv](https://drive.google.com/file/d/18f1srPV-Bwp_JH_t4_eujW1MgY1H3rZv/view?usp=drive_link))
2. Evaluar el rendimiento del modelo

```python
import pandas as pd
import requests

# Google Drive file ID for data.csv
file_id = '18f1srPV-Bwp_JH_t4_eujW1MgY1H3rZv'
url = f'https://drive.google.com/uc?id={file_id}'

def download_file_from_google_drive(id, destination):
    URL = "https://docs.google.com/uc?export=download"

    session = requests.Session()

    response = session.get(URL, params = { 'id' : id }, stream = True)
    token = get_confirm_token(response)

    if token:
        params = { 'id' : id, 'confirm' : token }
        response = session.get(URL, params = params, stream = True)

    save_response_content(response, destination)

def get_confirm_token(response):
    for key, value in response.cookies.items():
        if key.startswith('download_warning'):
            return value

    return None

def save_response_content(response, destination):
    CHUNK_SIZE = 32768

    with open(destination, "wb") as f:
        for chunk in response.iter_content(CHUNK_SIZE):
            if chunk:
                f.write(chunk)

# Specify the destination file name
destination = 'data.csv'

# Download the file
download_file_from_google_drive(file_id, destination)
print(f"'{destination}' downloaded successfully.")
```

# Challenge: Wine Quality data

## Objetivo:
Determinar la calidad del vino con base en su características

1. Aplicar el algoritmo de KNN a Wine Quality Data que se encuentra dentro del siguiente link 👇🏼

> https://archive.ics.uci.edu/ml/datasets/Wine+Quality

2. Evaluar el rendimiento del modelo

El objetivo de este desafío es construir un modelo de K-Nearest Neighbors (KNN) para predecir la calidad del vino tinto basándose en sus propiedades fisicoquímicas.

### Download and Load Wine Quality Dataset

```python
import pandas as pd
import requests

# URL for the winequality-red.csv dataset
wine_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'

# Download the dataset
response = requests.get(wine_url)
with open('winequality-red.csv', 'wb') as f:
    f.write(response.content)

print("'winequality-red.csv' downloaded successfully.")

# Load the dataset into a pandas DataFrame
df_wine = pd.read_csv('winequality-red.csv', sep=';')

# Display the first 5 rows of the DataFrame
display(df_wine.head())
```

Una vez descargado y cargado el dataset, mostramos las primeras filas para verificar que la carga fue exitosa y entender la estructura de los datos.

### Explore Wine Quality Dataset

### Exploración de Datos del Dataset de Calidad de Vino

```python
# Display basic information about the DataFrame
print(df_wine.info())

# Display descriptive statistics
print(df_wine.describe())
```

Los métodos `info()` y `describe()` nos proporcionan un resumen inicial del dataset. `info()` muestra el tipo de datos de cada columna y si hay valores nulos, mientras que `describe()` ofrece estadísticas descriptivas como la media, desviación estándar, valores mínimos y máximos, y los cuartiles.

### Prepare Data for KNN Model

### Preparación de Datos para el Modelo KNN

```python
# Separate features (X) and target (y)
X_wine = df_wine.drop('quality', axis=1)
y_wine = df_wine['quality']

print(f"Shape of features (X_wine): {X_wine.shape}")
print(f"Shape of target (y_wine): {y_wine.shape}")
```

Separamos el dataset en características (X) y la variable objetivo (y), que en este caso es la 'quality' del vino. Esto es un paso fundamental antes de entrenar cualquier modelo de aprendizaje automático.

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
# We'll use 80% for training and 20% for testing, with a random state for reproducibility
X_wine_train, X_wine_test, y_wine_train, y_wine_test = train_test_split(X_wine, y_wine, test_size=0.2, random_state=42)

print(f"Shape of X_wine_train: {X_wine_train.shape}")
print(f"Shape of X_wine_test: {X_wine_test.shape}")
print(f"Shape of y_wine_train: {y_wine_train.shape}")
print(f"Shape of y_wine_test: {y_wine_test.shape}")
```

Dividir los datos en conjuntos de entrenamiento y prueba es crucial para evaluar el rendimiento del modelo en datos no vistos. Usaremos el 80% para entrenar y el 20% para probar, con un estado aleatorio (`random_state`) para asegurar la reproducibilidad.

### Feature Scaling

### Escalado de Características

```python
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
scaler_wine = StandardScaler()

# Fit the scaler on the training data and transform both training and test data
X_wine_train_scaled = scaler_wine.fit_transform(X_wine_train)
X_wine_test_scaled = scaler_wine.transform(X_wine_test)

print("Features scaled successfully for both training and testing sets.")
print(f"Shape of scaled X_wine_train: {X_wine_train_scaled.shape}")
print(f"Shape of scaled X_wine_test: {X_wine_test_scaled.shape}")
```

El escalado de características es un paso importante para algoritmos basados en distancia como KNN. Utilizamos `StandardScaler` para estandarizar las características, transformándolas para que tengan una media de 0 y una desviación estándar de 1. Esto evita que las características con rangos de valores más grandes dominen el cálculo de distancias.

### Train and Evaluate KNN Model

### Entrenamiento y Evaluación del Modelo KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

# Initialize the KNN classifier with n_neighbors = 5
knn_wine = KNeighborsClassifier(n_neighbors=5)

# Train the model using the scaled training data
knn_wine.fit(X_wine_train_scaled, y_wine_train)

print("KNN model trained successfully.")

# Make predictions on the scaled test data
y_wine_pred = knn_wine.predict(X_wine_test_scaled)

print("Predictions made on the test set.")
```

Aquí inicializamos y entrenamos el modelo KNN con 5 vecinos (`n_neighbors=5`). Luego, usamos el modelo entrenado para hacer predicciones sobre el conjunto de prueba escalado (`X_wine_test_scaled`).

```python
# Evaluate the model performance
accuracy_wine = accuracy_score(y_wine_test, y_wine_pred)
print(f"Accuracy of KNN model on Wine Quality dataset: {accuracy_wine:.4f}")

# Display classification report
print("\nClassification Report:")
print(classification_report(y_wine_test, y_wine_pred))
```

Para evaluar el modelo, calculamos la precisión (`accuracy_score`) y generamos un `classification_report`. La precisión nos da una idea general de cuántas predicciones fueron correctas. El informe de clasificación proporciona métricas más detalladas como precisión (precision), exhaustividad (recall) y puntuación F1 (f1-score) para cada clase de calidad de vino, lo que nos ayuda a entender el rendimiento del modelo para cada nivel de calidad.
